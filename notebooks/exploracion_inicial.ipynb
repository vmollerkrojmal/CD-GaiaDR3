```python
from astroquery.vizier import Vizier
from astropy.table import Table
```

Descargar catálogo Córdoba-Durchmusterung

```python
# No limitar cantidad de columnas
Vizier.ROW_LIMIT = -1

# ID del catálogo Córdoba-Durchmusterung
catalog_id = "I/114/cd"

catalogs = Vizier.get_catalogs(catalog_id)
```

```python
len(catalogs)
```

```python
cd_table = catalogs[0]
```

```python
cd_table.info()
```

```python
display(cd_table)
```

Interpretar coordenadas originales

```python
from astropy.coordinates import SkyCoord, FK5
import astropy.units as u
import numpy as np
```

```python
coords_old = SkyCoord(
    ra=cd_table['RA1875'],
    dec=cd_table['DE1875'],
    unit=(u.hourangle, u.deg),
    frame=FK5(equinox="J1875")
)
```

Transformar a ICRS

```python
coords_icrs = coords_old.icrs
```

Agregar coordenadas ICRS al catálogo

```python
cd_table["ra_icrs"] = coords_icrs.ra.deg
cd_table["dec_icrs"] = coords_icrs.dec.deg
```

```python
cd_table
```

Tabla solo con ra y dec para match con Gaia

```python
cd_small = cd_table[["ra_icrs", "dec_icrs"]]
```

```python
cd_small.rename_columns(["ra_icrs", "dec_icrs"], ["ra", "dec"])
```

```python
cd_small.write("cd_catalog_icrs.csv", format="csv", overwrite=True)
```

```python
cd_small.write("cd_catalog_icrs.fits", format="fits", overwrite=True)
```
